```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
age_x=[25,26,27,29,33,32,35,28,35,39]
devop=[3525,3326,3627,3129,3633,3032,3115,3628,3435,3009]
javadop=[4525,5326,5627,5129,5633,5032,6115,6628,6435,6009]
```

```python
sorted_age_x = np.sort(age_x)
sorted_devop = np.sort(devop)
sorted_javadop = np.sort(javadop)
```

```python
sorted_devop
```

```python
plt.plot(sorted_age_x,sorted_devop,label="All devep", color = 'hotpink')
plt.plot(sorted_age_x,sorted_javadop,label="All Java ")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```

```python
plt.scatter(sorted_age_x,sorted_devop,label="All devep", color = 'hotpink')
plt.scatter(sorted_age_x,sorted_javadop,label="All Java ")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```

```python
plt.bar(sorted_age_x,sorted_devop,label="All devep")
plt.bar(sorted_age_x,sorted_javadop,label="All Java ")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.show()
```
