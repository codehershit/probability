```python
import statsmodels.api as sm
import numpy as np

import pandas as pd 
import matplotlib.pyplot as plt
np.random.seed(9823655)
```

```python
nsample=100
x=np.linspace(0,10,100)
x
X=0
```

```python
X=np.column_stack((x,x**2))
beta=np.array([1,0.1,10])
e=np.random.normal(size=nsample)
X.shape
```

```python
X=sm.add_constant(X)
X
```

```python
y=np.dot(X,beta)+e
```

```python
z=np.column_stack((X,y))
```

```python
z
```

```python
data = pd.DataFrame(z)
data.columns=['c','x','x2','Y']
```

```python
data[0:10]
```

```python
data.to_csv("stackeddata.csv", index=False)
```
